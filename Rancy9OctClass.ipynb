```python
#Now create lists with content
#Lists can contain values of any data types
#Lists can contain duplicate value

product1 = ["something", 100, 200, [10,20,30], True, 10.25]
product1
```

```python
#Learn indexing
superheros = ["Thor","IronMan", "WonderWoman", "Thanos"]
superheros[0]
```

```python
superheros[-1]
```

```python
#Traversing means accessing/visiting elements 1by1
for myhero in superheros:
    print(myhero)
```

```python
#Add new items to a list using append() method, it will add to the end
superheros.append("Gamora")
print(superheros)
```

```python
#Add new item at a desired position it does using insert()
superheros.insert(2,"Cap. America")
print(superheros)
```

```python
#how to add/merge a list into another list
# using extend() method
DC = ["batman","superman","aquaman"]
superheros.extend(DC)
print(superheros)
```

```python
#Remove element using remove()
superheros.remove("Thanos")
print(superheros)
```

```python
# removing using del keyword
del superheros[4]
print(superheros)
```

```python
# del keyword can be used to remove elements with slicing
del superheros[1:4]
print(superheros)
```

```python
superheros1= ["Thor","IronMan", "WonderWoman", "Thanos"]
DC = ["batman","superman","aquaman"]
```

```python
superheros1.extend(DC)
print(superheros1)
```

```python
#pop to remove last element
superheros1.pop()
print(superheros1)
```

```python
#pop works with indexes also
superheros1.pop(3)
print(superheros1)
```

```python
# List is mutable which means it is alterable
superheros1[0]="God of Thunder"
print(superheros1)
```

```python
# Sorting a list
superheros1.sort()
```

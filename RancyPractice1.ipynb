#Given an integer,n, perform the following conditional actions:
    <ul>
<li>If is odd, print Weird</li>
<li>If is even and in the inclusive range of 2 to 5, print Not Weird</li>
<li>If is even and in the inclusive range of 6 to 20, print Weird</li>
<li>If is even and greater than 20, print Not Weird</li>
    </ul>
</span>

```python
n=int(input("Enter String:"))
if n%2==0 and n>=2 and n<=5:
    print("Not Weird")
elif n%2==0 and n>=6 and n<=20:
    print("Weird")
elif n%2==0 and n>=20:
    print("Not Weird")
else:
    print("Weird")
```

Description
In a certain encrypted message which has information about the location(area, city), the characters are jumbled such that first character of the first word is followed by the first character of the second word, then it is followed by second character of the first word and so on

In other words, let’s say the location is bandra,mumbai
The encrypted message says ‘bmaunmdbraai’.
Sample Input:
bmaunmdbraai
Sample Output:
bandra,mumbai
Let’s say the size or length of the two words wouldn’t match then the smaller word is appended with # and then encrypted in the above format.

With this in mind write a code to identify the right location and print it as place,city.

```python
jumbledCity = input("jumbled place, city name:")
message1 = ""
message2 = ""
wordlength = len(jumbledCity)
print(wordlength)

i=0
while i<wordlength:
    if i%2==0:
        if jumbledCity[i]!="#":
            message1 += jumbledCity[i]
    else:
        if jumbledCity[i]!="#":
            message2 += jumbledCity[i]
    i+=1
print(message1, message2)
```
